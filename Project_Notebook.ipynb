```python
train_data['transmission'] = (
    train_data['transmission']
    .str.strip().str.lower()  # clean spaces and case
    .replace({
        'anual': 'manual',
        'manua': 'manual',
        'anua': 'manual',
        'automatic': 'automatic',
        'utomatic': 'automatic',
        'automati': 'automatic',
        'utomati': 'automatic',
        'semi-aut': 'semi-auto',
        'emi-auto': 'semi-auto',
        'emi-aut': 'semi-auto',
        'unknow': 'unknown',
        'nknown': 'unknown',
        'nknow': 'unknown'
    })
)
train_data['transmission'].value_counts()
```
